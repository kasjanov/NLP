##### Preprocess for models fitting

```python
assert True
sentences = []
num_news = 0
for line in tqdm_notebook(list_news):
    spls = pre.text_preprocess_light(line).split()
    sentences.append(spls)
    num_news += 1
    if num_news > 10000:
        break
```

```python
sentences = [i for i in sentences if len(i) > 2]
```

```python
sentences[0][:10]
```

```python
%%time
modelFT = FastText(sentences=sentences, vector_size=300, min_count=1, window=5, workers=2)
```

```python
%%time
modelW2V = Word2Vec(sentences=sentences, vector_size=300, window=5, min_count=1, workers=2)
```

```python
w2v_index = annoy.AnnoyIndex(300 ,'angular')
ft_index = annoy.AnnoyIndex(300 ,'angular')

counter = 0

for question in tqdm_notebook(sentences):
    n_w2v = 0
    n_ft = 0
    vector_w2v = np.zeros(300)
    vector_ft = np.zeros(300)
    for word in question:
        if word in modelW2V.wv:
            vector_w2v += modelW2V.wv[word]
            n_w2v += 1
        if word in modelFT.wv:
            vector_ft += modelFT.wv[word]
            n_ft += 1
    if n_w2v > 0:
        vector_w2v = vector_w2v / n_w2v
    if n_ft > 0:
        vector_ft = vector_ft / n_ft
    w2v_index.add_item(counter, vector_w2v)
    ft_index.add_item(counter, vector_ft)

    counter += 1

    if counter > 100000:
        break

w2v_index.build(10)
ft_index.build(10)
```

```python
def get_response(question, index, model, number_answers):
    question = pre.text_preprocess_light(question).split()
    vector = np.zeros(300)
    norm = 0
    for word in question:
        if word in model.wv:
            vector += model.wv[word]
            norm += 1
    if norm > 0:
        vector = vector / norm
    answers = index.get_nns_by_vector(vector, number_answers, )
    return [list_news[i] for i in answers]
```

```python
question = 'Экономическое развитие'
for answer_news in tqdm_notebook(get_response(question, w2v_index, modelW2V, 5)):
    print(f'{answer_news}\n')
```

```python
question = 'Экономическое развитие'
for answer_news in tqdm_notebook(get_response(question, ft_index, modelFT, 5)):
    print(f'{answer_news}\n')
```
